```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px 
import seaborn as sns

# %matplotlib inline - to ensure charts created are shown directly and not as pop ups

titanic_df = pd.read_csv('test.csv')
titanic_df
```

```python
# Drop a column that's mostly missing (deck: 688/891 missing)
titanic_df.drop(columns='Cabin', inplace=True)

# Input a reasonable numeric value to Age and Fare  column (median)
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Fare'] = titanic_df['Fare'].fillna(titanic_df['Fare'].median())
titanic_df.isnull().sum()
```

```python
titanic_df['Fare'] = titanic_df['Fare'].astype(float)
titanic_df['Age'] = titanic_df['Age'].astype(int)
```

```python
#  DEALING WITH THE AGE AND FARE OUTLIERS

Q1_age = titanic_df['Age'].quantile(0.25)
Q3_age = titanic_df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_age_bound = Q1_age - 1.5 * IQR_age
upper_age_bound = Q3_age + 1.5 * IQR_age

Q1_fare = titanic_df['Fare'].quantile(0.25)
Q3_fare = titanic_df['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_fare_bound = Q1_fare - 1.5 * IQR_fare
upper_fare_bound = Q3_fare + 1.5 * IQR_fare

print(f"Outlier bounds for Age: {lower_age_bound:.2f} to {upper_age_bound:.2f}")
print(f"Outlier bounds for Fare: {lower_fare_bound:.2f} to {upper_fare_bound:.2f}")

# Build boolean masks — each column checked against its OWN bounds
age_mask = (titanic_df['Age'] >= lower_age_bound) & (titanic_df['Age'] <= upper_age_bound)
fare_mask = (titanic_df['Fare'] >= lower_fare_bound) & (titanic_df['Fare'] <= upper_fare_bound)

# Combine masks, then filter once
combined_mask = age_mask & fare_mask
titanic_clean = titanic_df[combined_mask]

titanic_clean
```
